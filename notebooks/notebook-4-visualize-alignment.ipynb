# Now we can use the genomeview package in python to visualize this data.

```python
import genomeview
```

Let us define our data that we will pass to genomeview, genomeview needs the "pseudoalignments.bam" files from each of our quant folders generated by Kallisto

```python
datasets = {"High-Fat Diet Control 1":"/home/" + MIAMIID + "/test/kallisto/analyzed/SRR5017135_trimmed.fastq.gz_quant/pseudoalignments.bam",
            "High-Fat Diet Control 2":"/home/" + MIAMIID + "/test/kallisto/analyzed/SRR5017137_trimmed.fastq.gz_quant/pseudoalignments.bam",
            "High-Fat Diet Control 3":"/home/" + MIAMIID + "/test/kallisto/analyzed/SRR5017133_trimmed.fastq.gz_quant/pseudoalignments.bam",
            "High-fat Diet Tumor 1":"/home/" + MIAMIID + "/test/kallisto/analyzed/SRR5017132_trimmed.fastq.gz_quant/pseudoalignments.bam", 
            "High-fat Diet Tumor 2":"/home/" + MIAMIID + "/test/kallisto/analyzed/SRR5017128_trimmed.fastq.gz_quant/pseudoalignments.bam", 
            "High-fat Diet Tumor 3":"/home/" + MIAMIID + "/test/kallisto/analyzed/SRR5017138_trimmed.fastq.gz_quant/pseudoalignments.bam"}

reference = "/home/" + MIAMIID + "/test/genomes/Mus_musculus.GRCm38.dna.primary_assembly.fa"
chrom = "chr6"
start = 29060195
end = 29073877
```

```python
visualization = genomeview.visualize_data(datasets, chrom, start, end, reference)
```
